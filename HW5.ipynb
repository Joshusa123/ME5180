```python
import numpy as np
from numpy import sin,cos,pi
import matplotlib.pyplot as plt
from scipy.integrate import odeint
plt.style.use('fivethirtyeight')
```

```python
#Problem 1 - Define ODE for the rollercoaster motion

def cart_ode2(U, x):
   # Here U is a vector of the generalized coordinate and its derivative
   # U = [x, xdot]
    g = 9.81;
    w_2   =0;   #Rate at which parabolic track rotates (rad) Problem 2
    
   #Define function that is the derivative of U, so f = dU/dt = [xdot, xdotdot]
    funky = [U[1], 
             -(U[1]**2*U[0])/(1+U[0]**2)-((g-w_2**2)*U[0])/(1+U[0]**2)];
    
    return funky

def cart_ode3(U, x):
   # Here U is a vector of the generalized coordinate and its derivative
   # U = [x, xdot]
    g = 9.81;
    w_3   = 1;   #Rate at which parabolic track rotates (rad) Problem 3
    
   #Define function that is the derivative of U, so f = dU/dt = [xdot, xdotdot]
    funky = [U[1], 
             -(U[1]**2*U[0])/(1+U[0]**2)-((g-w_3**2)*U[0])/(1+U[0]**2)];
    
    return funky

def cart_ode4(U, x):
   # Here U is a vector of the generalized coordinate and its derivative
   # U = [x, xdot]
    g = 9.81;
    w_4   = 2;   #Rate at which parabolic track rotates (rad) Problem 4
    
   #Define function that is the derivative of U, so f = dU/dt = [xdot, xdotdot]
    funky = [U[1], 
             -(U[1]**2*U[0])/(1+U[0]**2)-((g-w_4**2)*U[0])/(1+U[0]**2)];
    
    return funky
```

```python
#Problem 2 - Parameters
x0_2  = 10;  #Initial Horizontal Position (x-position) on parabolic track (m)
dx0_2 = 0;   #Initial Horizontal Velocity (x-velocity) on parabolic track (m/s)
end_time2=12.5; # choose an end time that displays one full period

t2 = np.linspace(0,end_time2,500);
X2o = [x0_2,dx0_2]
X2 = odeint(cart_ode2, X2o, t2)

plt.xlabel("time (sec)")
plt.ylabel("Horizontal Displacement (x in meters)")
plt.title("Problem 2")
plt.plot(t2,X2[:,0],label=r'Nonlinear Numerical');
```

```python
#Problem 3 - Parameters
x0_3  = 3;  #Initial Horizontal Position (x-position) on parabolic track (m)
dx0_3 = 0;   #Initial Horizontal Velocity (x-velocity) on parabolic track (m/s)
end_time3=4.8; # choose an end time that displays one full period

t3 = np.linspace(0,end_time3,100);
X3o = [x0_3,dx0_3]
X3 = odeint(cart_ode3, X3o, t3)

plt.xlabel("time (sec)")
plt.ylabel("Horizontal Displacement (x in meters)")
plt.title("Problem 3")
plt.plot(t3,X3[:,0],label=r'Nonlinear Numerical');
```
