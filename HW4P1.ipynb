```python
import numpy as np
from scipy import linalg
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')

m = 0.1
k = 20
L = 1
g = 9.81

M = np.array([[m, m*L],[m*L, m*L**2]])
K = np.array([[2*k, 0],[0, m*L*g]])

w2, v = linalg.eig(K,M)

print(M)
print(K)
```

```python
print('natural frequencies:',w2**0.5)
print('eigenmodes:')
print(v/v[0,:])
```
